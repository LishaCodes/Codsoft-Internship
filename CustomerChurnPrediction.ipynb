```python
#importing necessary library 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif
```

```python
# Loading the dataset and displaying the first few rows
bank = pd.read_csv('Churn_Modelling.csv')
bank.head()
```

```python
# Displaying information about the dataset
bank.info()
```

```python
# Displaying basic statistical information about the dataset
bank.describe()
```

```python
# Checking for missing values in the dataset
bank.isnull().sum()
```

```python
# Checking for duplicated rows in the datase
bank.duplicated()
```

```python
# Dropping unnecessary columns: 'CustomerId', 'RowNumber', and 'Surname'
bank.drop(['CustomerId','RowNumber','Surname'],axis=1,inplace=True)
bank.head()
```

```python
# Creating a boxplot to visualize the distribution of numerical features
plt.figure(figsize=(15,8))
sns.boxplot(data=bank)
plt.show()
```

```python
# Creating a countplot to visualize the distribution of the target variable 'Exited'
sns.countplot(data=bank,x='Exited')
```

```python
# Creating countplots to visualize the distribution of categorical features
sns.countplot(data=bank,x='IsActiveMember')
```

```python
sns.countplot(data=bank,x='HasCrCard')
```

```python
sns.countplot(data=bank,x='Gender')
```

```python
# Displaying the count of different classes in the target variable 'Exited'
bank['Exited'].value_counts()
```

```python
# Encoding categorical variables 'Gender' and 'Geography' using LabelEncoder
le=LabelEncoder()
bank['Gender']=le.fit_transform(bank['Gender'])
bank['Geography']=le.fit_transform(bank['Geography'])

bank.head()
```

```python
# Calculating and visualizing the correlation matrix
correlation = bank.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Displaying the number of unique values in each column
bank.nunique()
```

```python
# Separating features (x) and target variable (y)
x = bank.drop('Exited', axis=1)
y = bank['Exited']
# Calculating mutual information scores for each feature with respect to the target variable
mutual_info_scores = mutual_info_classif(x, y)
# Creating a barplot to visualize mutual information scores for features
feature_info = pd.DataFrame({'Feature': x.columns, 'Mutual_Info_Score': mutual_info_scores})
plt.figure(figsize = (12,6))
sns.barplot(x='Mutual_Info_Score', y='Feature', data=feature_info, color='teal')
plt.title('Mutual Information Scores for Features')
plt.xlabel('Mutual Information Score')
plt.ylabel('Feature')
plt.show()
```

```python
# Sorting features based on mutual information scores and displaying the top features
sorted_features = feature_info.sort_values(by='Mutual_Info_Score', ascending=False)
top_features = sorted_features.head(6)
print(top_features)
```
